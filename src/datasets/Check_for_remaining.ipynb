```python
import pandas as pd
```

```python
extracted_data=pd.read_csv("../../rawdata/extracted_data.csv")
all_songs=pd.read_csv("../../rawdata/sirivennela_songs.csv")
```

```python
extracted_data
```

```python
all_songs
```

```python
def normalize(s):
    return (
        s.astype(str)
         .str.lower()
         .str.strip()
         .str.replace("_"," ")
    )
```

```python
extracted_norm = normalize(extracted_data["song_name"])
all_norm = normalize(all_songs["song_name"])
```

```python
extracted_norm[0]
```

```python
extracted_set = set(extracted_norm)
```

```python
pending_songs = all_songs[~all_norm.isin(extracted_set)]
```

```python
pending_songs
```

```python
pending_songs.to_csv("../../rawdata/pending_songsv1.csv")
```
