```python
import re

TELUGU_RE = re.compile(r'[\u0C00-\u0C7F]')

def has_telugu(s):
    return bool(TELUGU_RE.search(s))

def is_label(s):
    return ":" in s and len(s.split()) <= 3

def extract_song(text):
    lines = [l.strip() for l in text.splitlines() if l.strip()]

    song_name = None
    lyrics = []
    collect = False
    seen_telugu_header = False

    for i, line in enumerate(lines):

        low = line.lower()

        # Song name
        if low == "song:" and i + 1 < len(lines):
            song_name = lines[i + 1].strip()
            continue

        # Detect Telugu section
        if low == "telugu":
            seen_telugu_header = True
            continue

        # Start collecting after Telugu header when Telugu actually appears
        if seen_telugu_header and has_telugu(line):
            collect = True

        # Stop on English transliteration or metadata
        if collect:
            if not has_telugu(line):
                break
            if is_label(line):
                break
            lyrics.append(line)

    return song_name, "\n".join(lyrics).strip()
```

```python
data = []

for item in serper_scrapped_data_paths:
    dir_path, _, songs = item
    movie_name = os.path.basename(os.path.normpath(dir_path))

    for song in songs:
        file_path = os.path.join(base_path, movie_name, song)
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read()
            song_name,lyrics=extract_song(content)
            new_data = {
                "movie_name": movie_name,
                "song_name": song_name,
                "lyrics": lyrics
            }
            data.append(new_data)
            
```
