```python
from rapidfuzz import process, fuzz
import pandas as pd
import re
```

```python
def normalize(s):
    s = s.astype(str).str.lower().str.strip()
    s = s.str.replace("_", " ", regex=False)
    s = s.str.replace(r"[–—-]", " ", regex=True)
    s = s.str.replace(r"&", " and ", regex=True)
    s = s.str.replace(r"[^a-z0-9\s]", "", regex=True)
    s = s.str.replace(r"\s+", " ", regex=True).str.strip()
    return s
```

```python
pending_all_songs["song_norm"] = normalize(pending_all_songs["song_name"])
sirivennela_cleaned["song_norm"] = normalize(sirivennela_cleaned["song_name"])
```

```python
raw_norms = pending_all_songs["song_norm"].unique().tolist()
raw_norm_set = set(raw_norms)
```

```python
clean_norms = sirivennela_cleaned["song_norm"].unique().tolist()

unmatched = [c for c in clean_norms if c not in raw_norm_set]
```

```python
matches = []
for c in unmatched:
    best = process.extractOne(c, raw_norms, scorer=fuzz.token_set_ratio)
    if best and best[1] >= 90: 
        matches.append((c, best[0], best[1]))

match_df = pd.DataFrame(matches, columns=["clean_norm", "raw_norm", "score"])
match_df.sort_values("score", ascending=False).head(20)
```

```python
mapped_raw_norms = set(sirivennela_cleaned["song_norm"]) | set(match_df["raw_norm"])
```
