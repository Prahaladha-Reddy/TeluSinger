```python
serper_scrapped_data_paths=os.walk("../../lyrics_serper_tape")
```

```python
base_path="../../lyrics_serper_tape"
```

```python
def clean_lyrics(text: str) -> str:
    marker = "..."
    idx = text.rfind(marker)
    if idx != -1:
        return text[idx + len(marker):].lstrip()
    return text.strip()
```

```python
data = []

for item in serper_scrapped_data_paths:
    dir_path, _, songs = item
    movie_name = os.path.basename(os.path.normpath(dir_path))

    for song in songs:
        file_path = os.path.join(base_path, movie_name, song)
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read()
            content=clean_lyrics(content)
            new_data = {
                "movie_name": movie_name,
                "song_name": song[:-4],
                "lyrics": content
            }
            data.append(new_data)
```

```python
def parse_song(text: str):
    lines = text.strip().splitlines()

    song_name = None
    lyrics_start_idx = None

    for i, line in enumerate(lines):
        line = line.strip()

        if song_name is None and line and ":" not in line:
            song_name = line
            continue

        if song_name and line == "":
            lyrics_start_idx = i + 1
            break

    lyrics = "\n".join(lines[lyrics_start_idx:]).strip() if lyrics_start_idx else ""

    return song_name, lyrics
```

```python
new_data_to_add_to_db=[]
for row in brand_new_data.itertuples(index=False):
    movie = row.movie
    lyrics = row.telugu_lyrics
    song_name, lyrics=parse_song(lyrics)
    
    new_data = {
                "movie_name": movie,
                "song_name": song_name,
                "lyrics": lyrics
            }
    new_data_to_add_to_db.append(new_data)
    
    print(new_data)
```
